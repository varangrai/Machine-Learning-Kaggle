```python
drop_list1 = ['perimeter_mean','radius_mean','compactness_mean',
              'concave points_mean','radius_se','perimeter_se',
              'radius_worst','perimeter_worst','compactness_worst',
              'concave points_worst','compactness_se','concave points_se',
              'texture_worst','area_worst']
x_1 = x.drop(drop_list1,axis = 1 ) 
x_1.head()
```

```python
f,ax = plt.subplots(figsize=(14, 14))
sns.heatmap(x_1.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax);
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
```

### Splitting the Data into 70% train, 15% val and 15% test

```python
X_train, X_test, y_train, y_test =  train_test_split(x_1, y, test_size=0.3)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.5)
```

### Logistic Regression

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train,y_train)
predictions = lr.predict(X_val)
print('Accuracy in Valiadation: ',accuracy_score(y_val,predictions))
print('F1 Score Validation: ',f1_score(y_val,predictions,pos_label='M'),'\n Confusion Matrix \n')
cm = confusion_matrix(y_val,predictions)
sns.heatmap(cm,annot=True,fmt="d");
```

### XGboost

```python
xgb_clf = xgb.XGBClassifier()      
xgb_clf = xgb_clf.fit(X_train,y_train)
predictions=xgb_clf.predict(X_val)
ac = accuracy_score(y_val,predictions)
print('Accuracy is: ',accuracy_score(y_val,predictions))
print('F1 Score is: ',f1_score(y_val,predictions,pos_label='M'),'\n Confusion Matrix \n')
```

### Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictions=dtree.predict(X_val)
ac = accuracy_score(y_val,predictions)
print('Accuracy is: ',accuracy_score(y_val,predictions))
print('F1 Score is: ',f1_score(y_val,predictions,pos_label='M'),'\n Confusion Matrix \n')
cm = confusion_matrix(y_val,predictions)
sns.heatmap(cm,annot=True,fmt="d");
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
predictions=rfc.predict(X_val)
ac = accuracy_score(y_val,predictions)
print('Accuracy is: ',accuracy_score(y_val,predictions))
print('F1 Score is: ',f1_score(y_val,predictions,pos_label='M'),'\n Confusion Matrix \n')
cm = confusion_matrix(y_val,predictions)
sns.heatmap(cm,annot=True,fmt="d");
```

#### It can be seen that logistic regression and XGboost are our two best classifiers, let's test the trained model on unbiased test set.

```python
predictions = lr.predict(X_test)
print('Accuracy in Valiadation: ',accuracy_score(y_test,predictions))
print('F1 Score Validation: ',f1_score(y_test,predictions,pos_label='M'),'\n Confusion Matrix \n')
cm = confusion_matrix(y_test,predictions)
sns.heatmap(cm,annot=True,fmt="d");
```

```python
predictions=xgb_clf.predict(X_test)
ac = accuracy_score(y_test,predictions)
print('Accuracy is: ',accuracy_score(y_test,predictions))
print('F1 Score is: ',f1_score(y_test,predictions,pos_label='M'),'\n Confusion Matrix \n')
cm = confusion_matrix(y_test,predictions)
sns.heatmap(cm,annot=True,fmt="d");
```

#### Let's calculate the best features on basis of XGB classisfier

```python
from sklearn.feature_selection import RFECV
xgb_clf = xgb.XGBClassifier() 
rfecv = RFECV(estimator=xgb_clf, step=1, cv=5,scoring='accuracy')   
rfecv = rfecv.fit(X_train, y_train)
predictions = rfecv.predict(X_val)

print('Optimal number of features :', rfecv.n_features_)
print('Best features :', X_train.columns[rfecv.support_])
print('Accuracy is: ', accuracy_score(y_val,predictions))
```
